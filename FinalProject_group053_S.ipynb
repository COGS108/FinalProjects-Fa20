## Background and Prior Work

The National Basketball Association (NBA), is a professional men’s basketball league consisting of thirty different teams across the United States and Canada. Within the NBA Bubble, basketball players resided in various hotel rooms throughout the resort, practiced amongst their teams, and competed against each other all within the Walt Disney World Resort for an entire duration of four months (July - October), depending on how well the team performed in the regular season and playoffs (2). Generally in sports, when two teams play, the team that is playing on their home field/court tends to perform better and win more often. # Data Cleaning

First we clean our NBA historical DataFrame by dropping columns of data of stats that aren't too useful in determining home court advantage, and also unnecessary columns, such as game date and game status

```python
# Drop unnecessary columns
nba_games = nba_games.drop(labels=['GAME_DATE_EST', 'AST_home', 'REB_home', 'GAME_STATUS_TEXT', 'AST_away', 'REB_away',
                                   'TEAM_ID_home', 'TEAM_ID_away'], axis = 1)


# Replace home and away team IDs with the names of the teams using the kaggle_team_data DataFrame
def change_id(int):
    list_of_IDs = list(kaggle_team_data.get('TEAM_ID'))
    for index, ID in enumerate(list_of_IDs):
        if ID == int:
            return kaggle_team_data.iloc[index]['CITY'] + ' ' + kaggle_team_data.iloc[index]['NICKNAME']

# Apply change_id function to the Home_TEAM_ID and VISITOR_TEAM_ID columns
nba_games['HOME_TEAM_ID'] = nba_games['HOME_TEAM_ID'].apply(change_id)
nba_games['VISITOR_TEAM_ID'] = nba_games['VISITOR_TEAM_ID'].apply(change_id)

#Rename columns to Home_Team and Away_Team
nba_games = nba_games.rename(columns = {'HOME_TEAM_ID': 'Home_Team', 'VISITOR_TEAM_ID':'Away_Team'})
```

In the next two cells, we drop unnecessary columns in all three datasets, rename the '3 pointers made' column, and create a new column displaying whether or not the team is playing at Home using the MATCH UP column information.

```python
def set_home(match_up):
    """
    Function to be used to add a column about whether a team is Home or Away to main 3 CSV's.
    """
    if '@' in match_up:
        return False
    else:
        return True
    
# Adds a column to each DF called Home where values are either True or False
bubble_season_box = bubble_season_box.assign(Home = bubble_season_box['MATCH UP'].apply(set_home))
season_2019_box = season_2019_box.assign(Home = season_2019_box['MATCH UP'].apply(set_home))
bubble_playoff_box = bubble_playoff_box.assign(Home = bubble_playoff_box['MATCH UP'].apply(set_home))

# Rename the plus minus columns to get parity across DataFrames.
season_2019_box = season_2019_box.rename(columns={'#ERROR!': '+/-'})
bubble_playoff_box = bubble_playoff_box.rename(columns={'+/-\'': '+/-'})
bubble_season_box = bubble_season_box.rename(columns={'+/-\'': '+/-'})
```

It's important for us not to use any data from teams that didn't compete in the bubble. Therefore, we throw out any games from the earlier 2019 season that include a non-bubble team. We don't the keep the games where non-bubble teams played against teams that went to the bubble because it could skew our data.

```python
# List of teams in bubble
bubble_teams = list(bubble_season_box['TEAM'].unique())

def only_bubble_teams(match_up):
    """
    To be used on 2019 season to throw away any teams that didn't compete in the bubble.
    """
    if match_up[:3] in bubble_teams and match_up[-3:] in bubble_teams:
        return match_up
    else:
        return 0
    
season_2019_box = season_2019_box.assign(MATCH_UP = season_2019_box['MATCH UP'].apply(only_bubble_teams))

season_2019_box = season_2019_box[season_2019_box.get('MATCH_UP') != 0]

season_2019_box = season_2019_box.drop(columns = ['MATCH_UP'])
```

### Paired T Test Analysis

In order to determine whether or not NBA teams feel the same effect of home court advantage in the bubble as opposed to during the 2019 season before the bubble, we will compare the average win percentage of home games played before the bubble to the average win percentage of home games played during the bubble using a Paired T Test. This test is used to compare dependent samples, and since we will be testing the win percentages of the same teams before and during the bubble, this test is the most appropriate.

<b>Null Hypothesis:</b> ($\mu_1 = \mu_2$) The average home game win percentage before the bubble is <b>equal to</b> the average home game win percentage during the bubble.<br>
<b>Alternative Hypothesis:</b> ($\mu_1 \neq \mu_2$) The average home game win percentage before the bubble is <b>not equal to</b> the average home game win percentage during the bubble.<br>
<b>Confidence Level:</b> $\alpha = 0.05$. This means that there is 95% confidence that the conclusion of this test will be valid.<br>

```python
# extract two arrays containing each team's home game win percentage
# one array is for before the bubble and the other is for during the bubble
reg_pct_byTeam = regular_season_win_percentage[regular_season_win_percentage['Home'] == True]['Win_Percentage'].values
bubble_pct_byTeam = bubble_win_percentage[bubble_win_percentage['Home'] == True]['Win_Percentage'].values

# two-sided t-test for the null hypothesis that 2 related samples have equal average (expected) values
tStat, pVal = stats.ttest_rel(reg_pct_byTeam, bubble_pct_byTeam)
print(tStat, pVal)
```

<b>T-statistic:</b> $t = 1.10789933075676$.<br>
<b>P-value:</b> $p = 0.28043386344916393$.<br>
<b>Decision:</b> Since our p-value (0.2804) is <b>greater than</b> our alpha value (0.05), we fail to reject the null hypothesis.<br>
<b>Conclusion:</b> Our Paired T Test demonstrates that there is insufficient evidence to support the claim that the average win percentages differ between home games played in the bubble versus outside the bubble.<br>

Because of how +/- is calculated, we are going to throw out that statistic, because it doesn't really carry any significance.

We will keep any statistic that has a correlation value above 0.3 or below -0.3. While a correlation of 0.3 is somewhat weak, we did not expect any of the statistics to have a very large correlation coefficient, because basketball games are determined by many factors besides just one stat.

```python
# Here we get a list of stat names that had a correlation coeff above abs(0.3) in the either correlation dictionary.
stat_names = []

for key, val in correlation_coeff.items():
    if val >= abs(0.3):
        stat_names.append(key)

for key, val in bubble_correlation_coeff.items():
    if val >= abs(0.3):
        if key in stat_names:
            continue
        stat_names.append(key)

# Next, we create a list bubble_indicator, which will classify a value as being "Pre-Bubble" or "Bubble".
# The list corr_values is also created to hold the correlation values of each key seen in stat_names.
bubble_indicator = []
corr_values = []
for key in stat_names:
    if key == '+/-':
        continue
    corr_values.append(correlation_coeff[key])
    corr_values.append(bubble_correlation_coeff[key])
    bubble_indicator.append('Pre_Bubble')
    bubble_indicator.append('Bubble')

# The stat column is made by taking the stat_names and repeating the stats twice to accomodate for the bubble and
# non-bubble stats.
stat_col = []
for i in stat_names:
    if i == '+/-':
        continue
    stat_col.append(i)
    stat_col.append(i)
```

```python
# Function that will be used later to grab stats out of our datasets
def get_stat(made, attempted = []):
    """
    To be used on stats that have a Made and Attempted category or singular stats. Collects the totals of these statistics 
    over all games. Takes in two arrays, lists, or series of numbers for made and attempted. 
    Returns a list of [total, average made per game, total attempted, percentage] or just [total, average] for singular stats.
    Rounds all numbers to the 3 decimals places.
    """
    if len(attempted) == 0:
        totals = [sum(made), round(made.mean(), 3)]
    else:
        totals = [sum(made), round(made.mean(), 3), sum(attempted), round(sum(made/sum(attempted)), 3)]
    return totals
```

Now we perform a T-test on the stats at home in the bubble and the stats at home when out of the Bubble. This will tell us if there were any significant changes in any of the stats when teams played at home in the bubble versus when they played at home in the normal season. The same is done for the away stats and the results are stored in the dataframe, p_value_df.

```python
p_values_home = {}
p_values_away = {}

# Adds the t-test done on each stat to the corresponding dictionary with the stat as the key.
for col in stat_diff_2019.columns:
    p_values_home[col] = stats.ttest_rel(stats_home_2019[col], bub_stats_home[col])[1]
    p_values_away[col] = stats.ttest_rel(stats_away_2019[col], bub_stats_away[col])[1]
    

# Creates the df using the dictionaries of the p_values.
p_value_df = pd.DataFrame.from_dict(p_values_home, orient = 'index')
p_value_df = p_value_df.rename(columns = {0: 'Home_Pvalues'})
p_value_df = p_value_df.assign(Away_Pvalues = p_values_away.values())

# Sorts the df in increasng order using column 'Home_Pvalues'.
p_value_df = p_value_df.sort_values('Home_Pvalues')
p_value_df
```
