## How we quantified Environmental Impact 
* We quantified the environmental impact of our dishes (from the ingredients in them) with two general measures: water usage and green house gas emissions.

### Water Usage (when cultivating crops and producing animal products)
* We defined water usage as the water footprint. According to The Water Footprint Network, "The water footprint measures the amount of water used to produce each of the goods and services we use" (https://waterfootprint.org/en/water-footprint/what-is-water-footprint/). Specifically for our data on crop and animal product water usage in the United States, the measurement is meters cubed of water used per ton of product produced. The original dataset divided the water footprint into three types: green, blue, and grey. However, our research question does not warrant us to maintain this distinction so we thereby aggregated the three types into one weighted average water footprint for each food product.


### Green House gas emissions 
* We defined green house gas emissions as any emissions that are produced during the ingredient creation process. Our data reported the emissions at every level of production for the ingredient, but our question does not require this narrow of a scope. So, we summed all of these categories together and used the total for our analysis.

````python
#used to plot the 6 food groups 
def plt_formatter(l,t,ax,xlabel,ylabel,food_group):
    pal = sns.color_palette("Blues_r", len(l))
    #gives larger values a darker blue
    rank = t.argsort().argsort()
    ax1=sns.barplot(x=l, y=t, ax=ax,palette=np.array(pal[::-1])[rank])
    ax1.set_xlabel(xlabel,fontsize=15)
    ax1.set_ylabel(ylabel,fontsize=15)
    ax1.set_xticklabels(l, fontsize=15)
    ax1.set_title(food_group, fontsize=15)

    #change to min in df 
    ax1.set(ylim=(-1, 60))
    return ax
````

> * To analyze the animal and crop water footprint datasets, we used the .describe() method to gather basic descriptive statistics that elucidated which food product in either dataset had the highest water footprint (highest volume of water used to produce one ton of the food product).

Rather than assign each food product its own index (as was desired), the indices were each food productâ€™s blue, green, and grey water footprints. In other words, there were three observations for every one food product for both datasets (when we were hoping for one observation per food product). 

> * To get these data in usable formats, we had to generalize the specific green, blue, and grey water footprints to the weighted average water footprints for each product (by employing the .mean() method for the three types) in order to attain our goal of getting one observation for each food product. Once each of the food products' water footprints were averaged, we filtered the original dataset to only include the columns that would be conducive to answering our research question. For more detailed steps of how we got these data in usable formats, please reference the annotations below.

> * The other pre-processing steps that were required came after seeing the water footprint data distributions in the EDA section. Necessary transformations were made to remove the effects of two outlier observations that were not even food products. We also decided to add a column that categorized each food product into one of the major food groups per the ChooseMyPlate program mentioned above. For more specific information about the other pre-processing steps we took, please reference the annotations below as well.

> * The Original water footprint datasets for both crops and animal products come from the Global water footprint of production (1996â€“2005) table (Table 1) in Arjen Y. Hoekstra and Mesfin M. Mekonnen's research article titled "The water footprint of humanity" (https://doi.org/10.1073/pnas.1109936109).

> * Overall, the same methods described above were used to wrangle and clean the ingredients dataset that was applied to the water footprint datasets. The only difference had to do with how the food products were grouped between the emissions dataset and the water footprint datasets. Based on the parameters the group used to generalize the emissions observations to be more compatible with the ingredients dataset, we followed the same parameters for the water footprint datasets while trying to maintain its observations' detail. We then needed to filter this water footprint dataset to include data from only the United States of America because to be consistent with the scope of our research question.

````python
# select the state columns, which are relevant to our question, and save them to new sub df
USA_crop_wf = crop_wf.loc[:, 'Alabama':'Wyoming']      
````

Considering that this USA_crop_wf dataframe kept the crop_wf dataframe's original indexing where each crop's waterfootprint type (green, blue, and grey) had its own index instead of the crop having its own index, around 2/3 of the total 32,531 NaNs will be negligible after performing the following transformation:

````python
# pandas originally gave each crop's green, blue, and grey waterfootprint its own index 
# group every three indices into one index (just interested in total waterfootprint)
# calculate the average of those three values to get the crops' average weighted wf by state
USA_crop_wf = USA_crop_wf.groupby(USA_crop_wf.index // 3).mean()
````

The numerical indices don't tell us much, so we need to retrieve the FAOSTAT crop descriptions from crop_wf. Also since we're working with an isolated series from crop_wf, none of the corresponding values to these product descriptions will be affected by dropping the NaNs.

````python
# extract crop descriptions from original crop_wf dataframe and clean it up
Crops = crop_wf['Product description (FAOSTAT)']
Crops = Crops.dropna()

# reset index so they're not multiples of three in the crops series
crops_nu = Crops.reset_index(drop = True)
crops_nu
````

Now we need to add the crops series to the USA_crop_wf dataframe and perform the following transformations on the USA_crop_wf dataframe:

````python
# assign the column 'crops' to the USA_crop_wf dataframe with the above series
USA_crop_wf = USA_crop_wf.assign(Crops = crops_nu)

# add column with the totaled weighted averages for each crop in each state
USA_crop_wf['USA Weighted Average'] = USA_crop_wf.sum(axis=1)

# reduce the USA_crop_wf dataframe to series (only indices and weighted average data)
USA_crop_wf = USA_crop_wf[['Crops', 'USA Weighted Average']]

# groupby crops and aggregate their USA Weighted Averages by sum
USA_crop_wf = USA_crop_wf.groupby(['Crops']).agg('sum')

# check out how the above transformations have changed our USA_crop_wf dataframe
USA_crop_wf
````

Now we need to categorize the crops into their respective food groups.

````python
crop_list = (USA_crop_wf.index).to_list()
# adapted from Julia Jones' code
food_grp_list1 = []

veg_dict= ['Agave Fibres Nes','Anise, badian, fennel, corian.','Artichokes','Asparagus','Avocados','String beans',
           'Sugar beet','Cabbages and other brassicas','Carrots and turnips','Cassava','Cauliflowers and broccoli',
           'Chillies and peppers, dry', 'Chillies and peppers, green','Cinnamon (canella)','Cloves', 'Coffee, green',
           'Cucumbers and gherkins','Eggplants (aubergines)','Garlic','Ginger','Jute','Lettuce and chicory',
           'Maize', 'Maize, green','Okra','Olives','Onions (inc. shallots), green', 'Onions, dry','Peas, dry',
            'Peas, green', 'Pigeon peas','Pepper (Piper spp.)', 'Peppermint','Potatoes','Sweet potatoes',
           'Pumpkins, squash and gourds','Ramie','Roots and tubers nes','Sisal','Spices, nes','Spinach',
           'Taro (coco yam)','Tea','Tomatoes','Vanilla','Vegetables fresh nes','Vetches','Yams','Yautia (cocoyam)']

fruit_dict=['apples','apricots','bananas','berries','cherries','currants','dates','figs','fruit','grape','grapes',
            'lemons','mangoes','oranges','melon','melons','papayas','peaches','pears','plantains','plums','tangerines']


protein_dict=['almond','nut','pork','bean','carob','chick peas','cow peas','lentil','lupin','pistachio','cashew']

grain_dict=['barley','wheat','seed','cereal','fonio','hemp','hops','millet','grain','oat','pulse','quinoa','rice','rye', 
   'sorghum','triticale']
other_dict=['oil','sugar']
````

The code block below demonstrates how we assigned each crop a food group.

````python
#searches for element in the given dictionary, when a match is found end search
# and return true 
def fg(crop_list,xdict,element,food_grp_lst,group):
    for v in xdict:
        if v in element:
            food_grp_lst.append(group)
            return True
    return False
#go through crop list 
for element in crop_list:
    #see if element is a veg 
    x=fg(crop_list,veg_dict,element,food_grp_list1,'veg')
    #if not veg see if fruit
    if x==False:
        x=fg(crop_list,fruit_dict,element.lower(),food_grp_list1,'fruit')
    #if not v or f check protein
    if x==False:
        x=fg(crop_list,protein_dict,element.lower(),food_grp_list1,'protein')
    #and so on 
    if x==False:
        x=fg(crop_list,protein_dict,element.lower(),food_grp_list1,'grain')
    if x==False:
        x=fg(crop_list,other_dict,element.lower(),food_grp_list1,'other')
    #if no category works 
    if x==False:
        food_grp_list1.append('NA')
````

# Interpretation

We can see that one method(stats.f_oneway) outputted a p value that **is** significant, while the other two(anova_stat, (OLS) model) outputted p values that are **not** significant. Let's try to figure out why that is later on.


Right now we will now perform **Tukeyâ€™s honestly significantly differenced (HSD)** to determine the pairs of significantly different regions.

````python
# perform multiple pairwise comparison (Tukey's HSD)
# unequal sample size data, tukey_hsd uses Tukey-Kramer test
res = stat()
res.tukey_hsd(df=anova_table, res_var='sum', xfac_var='Region', anova_model='sum ~ C(Region)')
res.tukey_summary
````

The above results from Tukey's HSD suggests that all pairwise comarisons for regions do **not** reject the null hypothesis (p < 0.05) and indicates again that there are **no** statistically significant differences between the regional groups.

# Testing ANOVA assumptions
We are now going to check the ANOVA assumptions using test statitics and visual approaches.

````python
# QQ-plot
# res.anova_std_residuals are standardized residuals obtained from ANOVA (check above)
sm.qqplot(res.anova_std_residuals, line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()

# histogram
plt.hist(res.anova_model_out.resid, bins='auto', histtype='bar', ec='k') 
plt.xlabel("Residuals")
plt.ylabel('Frequency')
plt.show()
````

The standarized residuals do lie around the 45-degree line which suggests that the residuals are approximately normally distributed. However, the histogram distributions suggest that the residuals are right skewed.

So we can run a Shapiro-Wilk test which will be used to check the normal distribution of residuals.
**Null hypothesis: data is drawn from a nomral distribution**.

````python
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue)
````

We can see that the p value is significant (p < 0.05), therefore we can reject the null hypothesis and conclude that the data is **not** drawn from a normal distribution.

As the data is not drawn from normal distribution, we want to test the Homogeneity of variances. **Null hypothesis: samples from regions have equal variances.**

Since the data is not drawn from a normal distribution, we can use **Levene's test** to check the Homogeneity of variances.

````python
res = stat()
res.levene(df=anova_table, res_var='sum', xfac_var='Region')
res.levene_summary
````
