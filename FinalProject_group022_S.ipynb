### Combine June and November COVID-19 Data

In order to better understand how covid cases have evolved over time and make clearer comparison, we decide to merge the two dataframe. After we have merged the two dataframe, we found calculating the difference between the confirmed and death cases of the June and November dataset useful for future analysis. Finally, we wanted to find the degree covid has increased over time for each county. However, when we are calculating the ratio, we found a few cases with infinity value. We realized that there were several counties with 0 confirmed and death cases, which resulted in dividing the difference by zero. To solve this problem, we replaced all 0 confirmed and death cases in June with 1. In principle, when we replace a very small number with 0, the result will not be affected too significantly. Since confirmed cases must be whole numbers, 1 is indeed the smallest number we can replace 0 with. Then, we calculated the percent change in covid cases for each county by subtracting the November dataset from the June dataset and then divided by the June dataset. We stored both the difference and the ratio in our dataframe. 

```python
merged_months = corona_df2.merge(corona_June2, on = ['County', 'State', 'Lat','Long'], how = 'inner', suffixes=('_nov', '_june'))
merged_months.Deaths_june = merged_months.Deaths_june.apply(lambda x: 1 if x==0 else x)
merged_months.Confirmed_june = merged_months.Confirmed_june.apply(lambda x: 1 if x==0 else x)

#We obtained the difference between confirmed cases in June and November 
#Then we calculated the percent change in confirmed cases and assigned the value in a new column
merged_months = merged_months.assign(Confirmed_difference = merged_months['Confirmed_nov'] - merged_months['Confirmed_june'])
merged_months = merged_months.assign(Confirmed_ratio = merged_months['Confirmed_difference'] / merged_months['Confirmed_june'])

#We obtained the difference between death cases in June and November 
#Then we calculated the percent change in death cases and assigned the value in a new column
merged_months = merged_months.assign(Deaths_difference = merged_months['Deaths_nov'] - merged_months['Deaths_june'])
merged_months = merged_months.assign(Deaths_ratio = merged_months['Deaths_difference'] / merged_months['Deaths_june'])


merged_months = merged_months[['County', 'State', 'Lat','Long','Confirmed_difference','Confirmed_ratio','Deaths_difference',
                             'Deaths_ratio']]
merged_months

merged_months.head(10)
```

## Politics Data

For the politics dataset, we obtain the dataset of the 2016 presidential election. In this dataset, it contains plenty of columns like fraction of vote, fraction of race, income, total vote and so on. Before we could perform exploratory data analysis and data visualization, we first had to organize and clean the dataset. Because we are only interested in the political preference of each county, so we only need the column of fraction of republican vote and democrat vote. Moreover, total votes is also important because it is positively correlated with the county’s population. Having the number_vote column can reflect the actual population of the county. In the dataset, there are some counties with no republican vote, democrat vote and number of votes, which means those data are not useful to help us determine the political preference. Therefore, we set those fraction of votes to null and drop them. 


```python
new_election_county = list(parseData("usa-2016-presidential-election-by-county.json"))
```

```python
state_name = [d['fields']['state'] for d in new_election_county[0]]
county_name = [d['fields']['county'] for d in new_election_county[0]]
republican_vote = [d['fields']['rep16_frac'] if "rep16_frac" in d['fields'] else np.nan for d in new_election_county[0]]
democrate_vote = [d['fields']['dem16_frac'] if "dem16_frac" in d['fields'] else np.nan for d in new_election_county[0]]
number_vote = [d['fields']['votes'] if "votes" in d['fields'] else np.nan for d in new_election_county[0]]
new_tb = pd.DataFrame({'state': state_name,'county_name':county_name,
              "republican_vote": republican_vote,"democrate_vote":democrate_vote,"number_vote":number_vote})
```

```python
new_tb
```

In addition, in the counties_name column, some counties’ names have a suffix. As we dig into the column deeply, most counties' names have the suffix “county”, “city”, and “parish” then followed by the state name. In order to combine with the corona dataset, we need to clean the counties_name column. Otherwise, the coutnies_name column does not match the County column in corona_df2 and two dataframe can not merge together. Therefore, I wrote a function called update_county and filter out the name of the county without any suffix.


```python
def update_county(data):
    data = data.split()
    d_word = []
    for d in data:
        if d !='County,' and d != 'city,' and d !='Parish,':
            d_word.append(d)
            continue
        else: break
        break
    county = ' '.join(d_word)
    return county
new_tb.county_name = new_tb.county_name.apply(update_county)
```

## Combining Covid dataset with Politics dataset

In order to make comparison between political preferences and covid cases more clearly and visualize the dataset in later parts, we decided to merge the covid dataframe with the political preference. We merged the two dataframe according to their counties' names and the state’s name because some counties have the same name in different states. 


```python
combined_data = pd.merge(merged_months, new_tb, left_on=['County','State'], right_on=['county_name','state'])
combined_data = combined_data.dropna(axis = 0)
```

```python
def change_negative(datum):
    if datum < 0:
        return 0
    return datum
combined_data['Deaths_difference'] = combined_data['Deaths_difference'].apply(change_negative)
combined_data['Deaths_ratio'] = combined_data['Deaths_ratio'].apply(change_negative)
```

```python
combined_data = combined_data.drop(['Lat', 'Long', 'state', 'county_name'], axis=1)
combined_data
```

# Classification


For the machine learning part, we decide to do a classification to organize data to agreed-on categories. Since our question is how the political preference in each county affects the increasing rate of COVID positive cases in counties of the US, so the measure and prediction to the speed of spreading the virus are very important. 

In order to classify the speed of spreading virus, we decide to use the median of confirmed_ratio as a standard. The counties have lower confirmed_ratio than the standard will classify as 0 in the speed_of_virus column, which indicates that the counties have relative	ly low speed of spreading the virus. The counties have higher confirmed_ratio than the standard will classify as 1  in the speed_of_virus column, which indicates that the counties have relative	ly high speed of spreading the virus. Then we build a model to predict the speed_of_virus column not only enable us to classify the speed of spreading the virus in different counties but also find out the relationship between political preference and the speed of spreading.

The political preference of a county is mainly reflected in the fraction of republican vote and democrat vote in the 2016 presidential election. Therefore, the fraction of republican vote would be our main feature in the classification. And we do not need the fraction of democrat vote because the fraction of republican vote and fraction of democrat vote add up almost to 1 except a tiny fraction of liberty vote. So by adding both the fractions of democrat vote and republican vote will not help much. Moreover, the total vote of a county reflects the population of the county, which is also an important factor in the classification.


Firstly, I shuffle the whole dataframe. Then I split the first 2600 rows as the training data, and the rest as the testing data. Then I use republican_vote and total vote two features in the model and use logistic regression to train our model. Our prediction has 66% accuracy, which is not bad since we have 50 % of 1 and 50% of 0 in the speed_of_virus column. Moreover, we have the precision and recall scores both above 50% and BER is less than 0.5, which further shows the correlation between fraction of republican vote and speed of spreading the virus and convinces our belief political preference of a county is related to the county's speed of spreading the virus.



```python
median_ratio = combined_data.Confirmed_ratio.median()
def binary_ratio (data):
    if data >= median_ratio:
        return 1
    else:
        return 0
combined_data['speed_of_virus'] = combined_data.Confirmed_ratio.apply(binary_ratio)
temp_tb = combined_data.sample(frac=1).reset_index(drop=True) #random shuffle
```

```python
train = temp_tb[:2600]
valid = temp_tb[2600:]
y_train = train.speed_of_virus
y_valid = valid.speed_of_virus

initial = np.array([1] * len(train))
republican_vote_train = train.republican_vote.to_numpy()
num_vote_train = train.number_vote.to_numpy()
x_train = []
for i in range(len(republican_vote_train)):
    x_train.append([initial[i]] + [republican_vote_train[i]] + [np.log(num_vote_train[i])])
    
initial_valid = np.array([1] * len(valid))
republican_vote_valid = valid.republican_vote.to_numpy()
num_vote_valid = valid.number_vote.to_numpy()
x_valid = []
for i in range(len(republican_vote_valid)):
    x_valid.append([initial_valid[i]] + [republican_vote_valid[i]] + [np.log(num_vote_valid[i])])
```

```python
mod = linear_model.LogisticRegression(C=1.0)
mod.fit(x_train,y_train)
pred = mod.predict(x_valid)
```

```python
TP = sum(np.logical_and(pred, y_valid))
FP = sum(np.logical_and(pred, np.logical_not(y_valid)))
TN = sum(np.logical_and(np.logical_not(pred), np.logical_not(y_valid)))
FN = sum(np.logical_and(np.logical_not(pred), y_valid))
```

```python
TP,FP,TN,FN
```

```python
accuracy = (sum(pred == y_valid)/len(y_valid))
Precision = TP / (TP + FP)
Recall = TP /(TP + FN)
BER = 1-0.5*(TP / (TP + FN) + TN / (TN + FP))
print('Accuracy is: '+str(accuracy))
print('Precision is: '+str(Precision))
print('Recall is: '+str(Recall))
print('BER is: '+str(BER))
```
